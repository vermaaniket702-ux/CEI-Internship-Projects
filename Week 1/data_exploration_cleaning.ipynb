####  6 .Create a derived column (total_amount = price * quantity)
- "quantity" column is not given in the dataset so we derieve it through "rating_count" column.
- We use rating_count because it gives the approximate value of how many people bought the item.
- But we have derieved new "quantity" column as it should not contain any zeros, as total price will become zero.
- Then we derieved "total_price" column and do some formatting on that.

```python
df['quantity'] = df['ratings_count'].clip(lower=1) # to avoid zero quantity
df['total_price'] = (df['quantity'] * df['initial_price'])   # calculate total price
df['total_price'] = df['total_price'].apply(lambda x: f'₹{x:,.0f}') # format total price with currency symbol and commas
df.head(20)
```
